```python
import json
import pathlib
import re
```

```python
data_path = pathlib.Path("/export/usuarios_ml4ds/lbartolome/Repos/umd/corpus-building/corpus_scraping/strict/comunidad-madrid_strict_corpus.json")
data_path_edited = pathlib.Path("/export/usuarios_ml4ds/lbartolome/Repos/umd/corpus-building/corpus_scraping/strict/comunidad-madrid_strict_corpus.json")
```

```python
def extract_blocks_from_file(filename):
    with open(filename, 'r') as file:
        content = file.read()

    # Use regular expression to find all matches of {} blocks
    pattern = r'\{([^}]*)\}'
    matches = re.findall(pattern, content)

    # Extract the content inside each {} block and store in a list
    extracted_content = [match.strip() for match in matches]

    return extracted_content

# Call the function and store the extracted content in a list
extracted_content_list = extract_blocks_from_file(data_path)
```

```python
all_elements = []
for el in extracted_content_list:
    elements = el.split(', "') # Split the content of each {} block by ', "'
    elements_dict = {}
    for e in elements:
        key = e.split('": ')[0].strip('"')
        value = e.split('": ')[1].strip('"').replace('\\"', '"')
        elements_dict[key] = value
    all_elements.append(elements_dict)
with open(data_path_edited, "w") as json_file:
    json.dump(all_elements, json_file)
```
