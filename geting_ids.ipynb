```python
import pandas as pd
```

```python
previous_ids = "data/mind_runs/rosie/v1/src_filter_ids.txt"

with open(previous_ids, "r") as f:
    ids = f.read().splitlines()
ids[0]
```

```python
df = pd.read_parquet("data/mind_runs/rosie/results/results_topic_24_54_p1.parquet")
```

```python
new_ids = df.source_chunk_id.values.tolist()

# check if there is any id in new_ids that is in ids
print(any(i in ids for i in new_ids))

# other way round
print(any(i in new_ids for i in ids))

# combine unique ids
all_ids = list(set(ids) | set(new_ids))

# check if there was overlap
print(len(all_ids) == len(set(ids)) + len(set(new_ids)))  # False if there were duplicates

# save all ids to file, one per line
with open("data/mind_runs/rosie/results/src_filter_ids_tp24.txt",
            "w") as f:
    f.write("\n".join(all_ids))
```
